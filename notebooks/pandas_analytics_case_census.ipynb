# Pandas Analytics: case Census

```python
import pandas as pd
```

```python
url = 'https://raw.githubusercontent.com/HOGENT-Databases/DB3-Workshops/master/data/adult.csv'
adults = pd.read_csv(url)
print(adults.head(100))
```

```python
adults.count()
```

```python
adults[adults['native-country'].isnull()]
```

```python
adults = adults.drop(['fnlwgt','education','relationship'],axis=1)
adults.head()
```

```python
adults.groupby('native-country').count()
```

```python
adults.groupby('native-country')['age'].describe()
```

```python
adults[(adults['native-country'].str.strip() == 'Vietnam') & (adults['age'] >= 60)]
```

```python
def f(status):
    mapping = {}
    mapping['Divorced'] = 0
    mapping['Married-AF-spouse'] = 1
    mapping['Married-civ-spouse'] = 2
    mapping['Married-spouse-absent'] = 3
    mapping['Never-married'] = 4
    mapping['Separated'] = 5
    mapping['Widowed'] = 6
    status = status.strip()
    return mapping[status]

adults['marital-status'] = adults['marital-status'].map(f)
adults.head()
```

```python
adults.groupby(['age','gender'])['hours-per-week'].sum().unstack().fillna(0)
```

```python
# alternative solution:
import numpy as np
table = pd.pivot_table(adults, values='hours-per-week', index='age',columns='gender',aggfunc=np.sum)
table
```
